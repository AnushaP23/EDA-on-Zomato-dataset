```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

### Observation: Zomato maximum records or transactions are from India After that USA and then United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
## Converting it into a dataframe
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

### Observation:
1. When rating is between 4.5 to 4.9 -----> Excellent
2. When rating is between 4.0 to 4.4 -----> Very Good
3. When rating is between 3.5 to 3.9 -----> Good
4. When rating is between 2.5 to 3.4 -----> Average
5. When rating is between 1.8 to 2.4 -----> Poor

```python
ratings.head()
```

### Bar-plot between Aggregate rating and Rating count

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,palette='husl')
plt.show()
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color')
```

### Observation: AS you can see the colors are not mapped let's map it by using pallete

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','brown'])
plt.show()
```

### Observation:
1. Not Rated count is very high
2. Maximum number of rating is between 2.5 to 3.4

### Count-plot
Basically used for plotting categorical variable

```python
plt.figure(figsize=(6,6))
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','brown'])
plt.show()
```
